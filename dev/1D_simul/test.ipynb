```python
import numpy as np
import matplotlib.pyplot as plt
import geone
```

```python
mask = np.zeros([3, 3])
mask[1, 1] = 1
```

```python
np.random.seed(1312)
n = 40
cm = geone.covModel.CovModel2D(elem=[("spherical", {"w":1, "r":[100, 50]})])
plt.imshow(mask, extent=[0, 300, 0, 300], alpha=.4)
x = np.random.random(size=n)*300
y = np.random.random(size=n)*300
v = np.random.random(size= n)
plt.scatter(x, y, c=v)
xp = np.array([x, y]).T
```

```python
res = geone.geosclassicinterface.simulate2D(cm, (3, 3), (100, 100), x=xp, v=v, mask=mask, verbose=2, add_data_point_to_mask=True, nthreads=1, nreal=100)["image"].val[:, 0, 1, 1]
```

```python
plt.hist(res)
```

```python
res2 = geone.geosclassicinterface.simulate2D(cm, (3, 3), (100, 100), x=xp, v=v, mask=mask, verbose=2, add_data_point_to_mask=False, nreal=10)["image"].val[:, 0, 1, 1]
```

```python
plt.hist(res2)
```

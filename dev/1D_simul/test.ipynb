```python
import numpy as np
import matplotlib.pyplot as plt
import geone
```

```python
np.random.seed(1312)
n = 40
cm = geone.covModel.CovModel2D(elem=[("spherical", {"w":1, "r":[100, 200]})])

# plt.grid()
x = np.random.random(size=n)*300
y = np.random.random(size=n)*300
v = np.random.random(size= n)
plt.scatter(x, y, c=v)
xp = np.array([x, y]).T
plt.scatter(150, 150, c="r")
plt.xlim(0, 300)
plt.ylim(0, 300)
plt.colorbar()
```

```python
geone.grf.grf2D(cm, (1, 1), (1, 1), (149.5, 149.5), x=xp, v=v)
```

```python
res = geone.covModel.sgs(x=xp, v=v, xu=np.array([[150, 150]]), cov_model=cm, nreal=1, method="ordinary_kriging")
```
